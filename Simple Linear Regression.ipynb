# Simple Linear Regression in TensorFlow

This very well known model is a good way to start your exploration in TensorFlow. <br/>

It is described by the following equation:

$$Y = aX + b$$

Where Y is the dependent and X is the independent variable. Our task is to adjust the parameters a — “slope” and b — “intercept” so that we 

best describe the data using a line.

For our example, let’s find out how eating burgers affect your resting heart rate. The data will be simulated, so no conclusions, please! 

Our data represents the average number of burgers eaten per day.

```python
X = np.random.rand(100).astype(np.float32)
```

```python
a  = 50
b = 40
Y = a * X + b
```

```python
plt.plot(X,Y)
```

```python
Y = np.vectorize(lambda y: y + np.random.normal(loc=0.0, scale=0.05))(Y)
a_var = tf.Variable(1.0)
b_var = tf.Variable(1.0)
y_var = a_var * X + b_var
```

```python
loss = tf.reduce_mean(tf.square(y_var-Y))
```

```python
optimizer = tf.train.GradientDescentOptimizer(0.5)
train = optimizer.minimize(loss)
```
